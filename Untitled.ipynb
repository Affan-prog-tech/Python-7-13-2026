```python
import pandas as pd
import numpy as np

customer_ID= [f"CUST{i}" for i in range(1,501)]
Category= np.random.choice(["Electronics", "Clothing", "Books", "Food", "Sports"],500)
Quantity= np.random.randint(1,11,500)
UnitPrice= np.random.randint(100,5000,500)

df= pd.DataFrame({
    "Customer ID":customer_ID,
    "Category":Category,
    "Quantity":Quantity,
    "Unit Price":UnitPrice
})

df.head()
```

```python
df["TotalPrice"]=Quantity * UnitPrice
df.head()
```

```python
Sales=df["Quantity"].sum()
print("Total Sales:",Sales)
```

```python
CategoryWise_Sales= df.groupby("Category")["Quantity"].sum()

print(CategoryWise_Sales)
```

```python
Highest_Purchase= df.loc[df["TotalPrice"].idxmax()]

print(Highest_Purchase)
```

```python
Average_purchase = df["TotalPrice"].mean()

print(Average_purchase)
```
